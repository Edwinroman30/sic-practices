```python
## Definicin de pila como clase

class Pila:

    def __init__(self):
        self.pila = []

    def esta_vacia(self):
        return len(self.pila) == 0

    def apilar(self, elemento):
        self.pila.append(elemento)

    def sacar(self):
        return None if self.esta_vacia(self.pila) else self.pila.pop()
```

```python
## Definicin de pila como clase

class Pila:

    def __init__(self):
        self.pila = []

    def __str__(self):
        pass

    def esta_vacia(self):
        return len(self.pila) == 0

    def apilar(self, elemento):
        self.pila.append(elemento)

    def sacar(self):
        return None if self.esta_vacia() else self.pila.pop()

    def peek(self): # vistazo/hojeada
        return None if self.esta_vacia() else self.pila[-1]
```

```python
stack = Pila()

stack.apilar(5)
stack.apilar(7)

stack.peek()
```

```python
def checar_parentesis(expr):
    aperturas = "[{("
    cierres   = "]})"
    pila = Pila() # Instanciamos de la clase Pila

    for char in expr:
        if char in aperturas:
            pila.apilar(char)
        elif char in cierres:
            if pila.esta_vacia():
                return False
            if aperturas.index(pila.sacar()) != cierres.index(char):
                return False
    return pila.esta_vacia()
```

# Unit 23: Colas

```python
def secuencia_josephus(n, k):
    secuencia = []
    cola = []
    for i in range(1, n + 1):
        cola.append(i)
    j = 1
    while len(cola) > 1:
        item = cola.pop(0)
        if j % k == 0:
            secuencia.append(item)
        else:
            cola.append(item)
        j += 1
    secuencia.append(cola.pop(0))
    return secuencia
```

```python
N = int(input("Ingrese el nmero de personas: "))
K = int(input("Ingrese el nmero de posiciones que se desan saltar : "))
print(secuencia_josephus(N, K))
```

```python
class Cola:
    def __init__(self):
        self.cola=[]

    def esta_vacio(self):
        return len(self.cola) == 0

    def encolar(self, item):
        self.cola.append(item)

    def decolar(self):
        return None if self.esta_vacio() else self.cola.pop(0)

    def size(self):
        return len(self.cola)
```

```python
cola = Cola()
cola.encolar("A")
cola.encolar("B")
print(cola.decolar())
cola.encolar("C")
print(cola.decolar())
print(cola.decolar())
print(cola.decolar())
```

```python
## Solucin al problema de Josephus con la clase Cola

def josephus(n, k):
    secuencia = []
    cola = Cola()
    for i in range(1, n + 1):
        cola.encolar(i)
    j = 1
    while cola.size() > 1:
        item = cola.decolar()
        if j % k == 0:
            secuencia.append(item)
        else:
            cola.encolar(item)
        j += 1
    secuencia.append(cola.decolar())
    return secuencia
```

```python
N = int(input("Indique el numero de personas(N): "))
K = int(input("Indique el numero a ser saltado(K): "))
print(josephus(N, K))
```
