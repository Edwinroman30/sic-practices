```python
# Espacio para desarrollar el código requerido, recuerde agregar algunos
# comentarios breves sobre su funcionamiento.

#1. Explique brevemente y en forma clara cómo funciona el algoritmo de ordenamiento de burbuja.
''' 
Bubble Sort o método de ordenamiento de burbuja es un algoritmo de ordenamiento cuyo funcionamiento coinsiste en ordenar los elementos de un 
conjunto (Lista,Array,etc) intercambiando las posiciones adyacentes de menor a mayor. Este proceso de intercambios entre elementos adyacente 
es repetido hasta que todos los elementos queden ordenados. 
'''

#2. Desarrollo un función donde se use el ordenamiento de burbuja aplicada a una lista de valores enteros.

S = [50, 30, 40, 10, 20]

def bubble_sort(values_set : list):
    length = len(values_set)
    
    for elem in range(length):  # Iterar sobre la cantidad de elementos para asegurar que todos los elementos sean comparados.
        for inx_val in range(length - 1):
            if values_set[inx_val] > values_set[inx_val + 1]:
                values_set[inx_val], values_set[inx_val + 1] = values_set[inx_val + 1], values_set[inx_val] # values_set interchanging

    print(values_set)

bubble_sort(S)
```

```python
# 1. Explique brevemente el funcionamiento del algoritmo Quick Sort.
''' 
Quicksort es un algorithm tipo de ordenamiento de valores en un conjunto. La manera en la que este opera es tomando un valor pivote o referencia
para realizar comparaciones con la lista completa. Al final este ordena a la izquierda todos los valores menores a dicho punto de referencia y a la 
derecha del mismo aquellos valores mayores al valor referencia pivote.

El valor del pivote puede ser el primero o último valor de la lista, al final el algoritmo realizará este proceso de manera recursiva, siendo 
la estrategía de particionamiento fundamentar para lograr este ordenamiento left and right alrededor del valor pivote y que por cierto, este valor
pivote irá cambiando de derecha a izquierda, repitiendo el mismo proceso hasta que todo quede organizado.
'''

''' 
Veamos un ejemplo:

Nota: Profe disculpe si esta en Ingles, pero es un pequeño exercicio que hago para poder aprender más sobre el idioma.

S = [50, 30, 40, 10, 20]

Pivot values = 20
[10] # Lower than pivot ridding from left to right
[50, 30, 40] # Highter than the pivot from left to right

[10, 20, 50, 30, 40] # After partitioning

The process did before it is call partitioning, so important in QuickSort.

Left side Pivot
Pivot values of the left side (P = 20)
Now the lenght of the left side pivot is ONE and also it is already sorted.
The left side is sorted. 


Now lets evaluete the Right side after the Pivot.

[10, 20, 50, 30 , 40] #First partition process done!
     P

Sub_pivot = 40
Now we apply again partitioning and the new pivot (Sub_pivot) is compare with the same process with the values
of the first pivot right side list, resulting in:

[10, 20, 30, 40, 50]

Both sides are sorted, left and right sides.
'''


# Una simple función, para intercambiar valores dada la posición en una lista.
def swap(list_elems: str, currentIdx: int, targetIdx):
    ''' Una simple función, para intercambiar valores dada la posición en una lista. '''
    length = len(list_elems)
    if (0 > currentIdx or length < currentIdx) or (0 > targetIdx or length < targetIdx): # Valido que las posiciones a cambiar esten en el rango validao del iterable busqueda.
        raise Exception("Incorrecto, ha especificado valores que estan fuera de rango...")
    
    list_elems[currentIdx], list_elems[targetIdx] = list_elems[targetIdx], list_elems[currentIdx] # Efectuación del deslizamiento o cambios.

def partition_func(sub_set: list, start:int, end:int):
    ''' Should return the last position taken by the Pivot, when all was travel.'''
    pivot_value = sub_set[end]
    pivot_idx = start #@
    i = start
    
    for x in range(start, end):
              
        # If the pivot is highter move it to the left
        if pivot_value > sub_set[i]:# if the current index is lower thant pivot
            swap(sub_set, pivot_idx, i)
            pivot_idx += 1
            
        i += 1 #increse the travelsal index.
    
    if pivot_value == sub_set[i]: # Cuando el indice de exploracion coincida con el valor del pivot, puede realizar el cambio que posiciona al pivot en el centro, luego de haber evaluado todo los elementos anteriores del pivot.
        swap(sub_set, pivot_idx, i) # Aún así, el condicional anterior puede ser omitido.
    
    return pivot_idx

def quick_sort(sub_set: list, start:int, end:int):
    
    if len(sub_set) == 1: # Caso cuando llega a un extremo, osea unico elemento por lo cual esta ordenado.
        return
    
    if start < end:
        pivot = partition_func(sub_set, start, end)
        quick_sort(sub_set, start, pivot -1)
        quick_sort(sub_set, pivot+1, end)


s = [7, 1, 3, 5, 2, 6, 4]

print("Before sorted: ", s)
quick_sort(s, 0, len(s) - 1)
print("Sorted by Quick Sort", s)
```
