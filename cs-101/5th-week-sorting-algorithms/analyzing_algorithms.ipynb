An algorithm is a set of steps (finite steps) that are carried out to solve a certain situation. As programmers, we can solve a problems with in different ways (multiples algorithms) and when we have those different solutions into the table, as engineers, raise the questions: Why the algorithm A is faster than the algorithm B? What is the best one, the most shorten in terms of line of code? Is it highly coupled to the programming language?.... 

All these sorts of questions are related with-in one word and a technique: `Effitiency` and `Analyze of Algorithms`.

Base on how the algorithm behave when solving the problem, we always gona have certain patterns to describe the algorithm behavior.
    - Best case escenario.
    - Avarege case.
    - Worst case (That always will be our reference the referring to it).

> How do we performe algorithm analyses, base on what?

For responding that question, in algotrithm analyzes we have certain predefined principles called `Primitive Rules` that describe a notation for expressing: how many steps are done by our algorithm and take it as a reference (Worst case time).

##### Primitive Operations:
- Assignment.
- Calculation.
- Indexing.
- Return/Calling.

Each one of the operations listed above must be counted to determine the complexity in term of time of the algorithm. (Each operation identified represent a unit of time).

Let's see an example from the Lalitha Natraj's video:

```python
def sum_until_n(N : int):
    sum_variable = 0 # (Assignment) = 1
    
    # In the for loop:
    # (Assignment of i) = 1; 
    # (Calculation) when comparing if i is lower than the give N = n + 1; (Plus one because the for loop make an additional incremental for stopping.)
    # (Assignment) + (Operation) when incrementing i++ internaly = 2n; 
    for i in range( N ):
        sum_until_n = sum_until_n + 1 # (Assignment) + (Operation) when incrementing and assigning = 2n; 

    return sum_until_n # (Return) = 1

''' Final time complexity of the algorithm: ''' 
# total_time_taken(n) = 1 + 1 + (n + 1) + 2n + 2n + 1
# total_time_taken(n) = 5n + 4
```
