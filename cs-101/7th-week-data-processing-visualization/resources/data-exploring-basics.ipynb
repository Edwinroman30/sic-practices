```python
import pandas as pd
import matplotlib.pyplot as plt

ted_data = pd.read_csv("../data/ted_main.csv")
ted_data.head()
```

```python
#It is the most viewed talk.
f1 = ted_data.loc[ ted_data["views"] == ted_data["views"].max() , ["title", "views"]]
f1

# Now let compare if it is the same than the most commented.
print("Now let compare if it is the same than the most commented.")
f2 = ted_data.loc[ ted_data["comments"] == ted_data["comments"].max(), ["title", "views"] ]
f2
#plt.show()
```
