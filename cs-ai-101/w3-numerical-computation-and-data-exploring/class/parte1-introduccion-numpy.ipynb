## 1.2. Fundamentos de NumPy Array

```python
import numpy as np # Mejor legibilidad

arr1 = np.array([1,2,3,4,5]) # Usan operaciones matemáticas para todo el bloque de datos.
arr2 = np.array([1,2,3,4,5])


print(id(arr1))
print(id(arr2))
id(arr1) == id(arr2) # Dir en memoria
```

```python
arr3=arr1 # Copia por referencia

print(id(arr1))
print(id(arr3))
id(arr1) == id(arr3)
```

```python
arr3 = arr1.copy() # Copia por valor

print(id(arr1))
print(id(arr3))
id(arr1) == id(arr3)
```

```python
print(len(arr1))
print(arr1.size)
```
